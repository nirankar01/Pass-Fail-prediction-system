

# LOGISTC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
#input hours
x=[[1],[2],[3],[4],[5],[6]]
#output fail->0,pass->1
y=[0,0,0,1,1,1]

# create model
model = LogisticRegression()
# train model
model.fit(x,y)

# input value
hours = float(input("Enter hours to study of Student "))

# predict model
predictOutput = model.predict([[hours]])
if predictOutput==1:
    print("Student is pass")
else:
    print("Student is fail")
```

```python
import joblib
```

```python
joblib.dump(model,"model.pkl")
print("Successfully save model")
```
